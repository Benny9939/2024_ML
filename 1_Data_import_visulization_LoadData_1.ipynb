### Shear a dataset using the URL from github


```python
import pandas as pd

output_path = './Salary_Data.csv'
path = output_path
df = pd.read_csv(output_path)
display(df.head())
```

```python
x = df["YearsExperience"]
y = df["Salary"]
```

```python
# download the chinese font
url="https://github.com/GrandmaCan/ML/raw/main/Resgression/ChineseFont.ttf" 
# output_path = '/content/drive/MyDrive/Colab_Notebooks/linear_regression/ChineseFont.ttf'
output_path = './ChineseFont.ttf'
wget.download(url, out=output_path)
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")

plt.scatter(x, y, marker="x", color="red")
plt.title("年資-薪水")
plt.xlabel("年資")
plt.ylabel("月薪(千)")
plt.show()
```

### Linear Regresion 

```python
y_pred = 10*x + 20
plt.plot(x, y_pred, color="blue", label="預測線")
plt.scatter(x, y, marker="x", color="red", label="真實數據")
plt.title("年資-薪水")
plt.xlabel("年資")
plt.ylabel("月薪(千)")
plt.xlim([0, 12])
plt.ylim([10, 140])
plt.legend()
plt.show()
```

```python
def plot_predic(w, b):
    y_pred = w*x + b  
    plt.plot(x, y_pred, color="blue", label="預測線")
    plt.scatter(x, y, marker="x", color="red", label="真實數據")
    plt.title("年資-薪水")
    plt.xlabel("年資")
    plt.ylabel("月薪(千)")
    plt.xlim([0, 12])
    plt.ylim([10, 140])
    plt.legend()
    plt.show()
    return 
w=10
b=20
plot_predic(w, b)
```

```python
from ipywidgets import interact

interact(plot_predic, w=(-100, 100, 2), b=(-100, 100, 1))
```
